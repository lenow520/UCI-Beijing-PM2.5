# Data preprocessing

```
PRSA.isnull().sum()
```

```
#Remove null values
PRSA.dropna(inplace=True)
PRSA.isnull().sum()
```

```
PRSA
```

```
#Remove useless column(s)
PRSA=PRSA.drop(['No'],axis=1)
```

```
PRSA['cbwd']=PRSA['cbwd'].astype('category')
PRSA.info()
```

```
#Encoding the categorical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
PRSA['cbwd']=le.fit_transform(PRSA['cbwd'])
PRSA
```

```
from sklearn import preprocessing 
col_names=['DEWP','TEMP','PRES','Iws','Is','Ir']
scaler=preprocessing.MinMaxScaler().fit(PRSA[col_names].values)
PRSA[col_names]=scaler.transform(PRSA[col_names].values)
PRSA
```

```
#Binning
bins = pd.cut(PRSA['pm2.5'], 5,labels=range(5)) 
bins
print(pd.value_counts(bins))
```

```
PRSA['pm2.5']=bins
PRSA.info()
```

# Feature selection

```
#select k best
from sklearn.feature_selection import SelectKBest, chi2
X=PRSA.drop(['pm2.5'],axis=1)
y = PRSA['pm2.5']
selector = SelectKBest(chi2,k=4)
selector.fit(X,bins)
X_new=selector.transform(X)
print(X.columns[selector.get_support(indices=True)])
```

```
import pandas as pd
from sklearn.model_selection import train_test_split
#split data into train and test sets
selected_col=['month', 'day', 'cbwd', 'Iws']
X=PRSA[selected_col]

X_train,X_test,y_train,y_test=train_test_split(X,bins,test_size=0.2,random_state=100)
```

# Classification

```
#Random forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100,max_depth=4,random_state=0)
clf.fit(X_train,y_train)
score=clf.score(X_test,y_test)
print("RandomForestClassifier:",score)
```

```
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logic = LogisticRegression(random_state=0)
logic.fit(X_train,y_train)
score=logic.score(X_test,y_test)
print("LogisticRegression:",score)
```

```
#Baye's classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
score=gnb.score(X_test,y_test)
print("GaussianNB:",score)
```

```
#ANN
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(X_train,y_train)
score=mlp.score(X_test,y_test)
print("ANN:",score)
```

```
# KMeans Algorithm
from sklearn import cluster
X=PRSA.drop(['pm2.5'],axis=1)
kmeans_fit = cluster.KMeans(n_clusters = 5).fit(X)
```

```
cluster_labels = kmeans_fit.labels_
print("Result:")
print(cluster_labels)
```
